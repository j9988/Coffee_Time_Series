### Moving Average for Smoothing

```python
df_ma= df_cleaned.copy()
df_ma['gen_ma7'] = df_ma['Total Generated (MW)'].rolling(window=7).mean()
df_ma['demand_ma7'] = df_ma['Maximum Demand (MW)'].rolling(window=7).mean()

sns.lineplot(data=df_ma, x='Date', y='gen_ma7', label='7-day Avg Generation')
sns.lineplot(data=df_ma, x='Date', y='demand_ma7', label='7-day Avg Demand')
plt.title("7-Day Moving Averages")
plt.show()
```

This code calculates and visualizes 7-day moving averages of electricity generation and demand to smooth short-term fluctuations and reveal underlying trends over time. The 7-day moving average of demand has a sharp spike around 2016, suggesting an unusual and significant short-term surge in power needs at that time. Moreover, generation and demand averages show a general upward trend over the years, with generation usually staying above demand, but the 2016 demand spike narrows this gap drastically.
